```python
import cv2
import numpy as np
import matplotlib.pylab as plt

#--① 이미지 그레이스케일로 읽기 및 출력
# 이미지를 그레이스케일로 읽어와 'img' 변수에 저장하고, 이를 화면에 출력
img = cv2.imread('../img/mountain.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('img', img)  # 읽은 이미지를 창에 표시

#--② 히스토그램 계산 및 그리기
# 그레이스케일 이미지의 히스토그램을 계산
hist = cv2.calcHist([img], [0], None, [256], [0, 255])
plt.plot(hist)  # 히스토그램을 그래프로 출력

#--③ 히스토그램의 shape 출력
# 히스토그램의 shape를 출력, (256,1)로 출력됨
print("hist.shape:", hist.shape)

#--④ 히스토그램 총 합계와 이미지의 크기 출력
# 히스토그램의 총 합계와 원본 이미지의 크기를 출력
print("hist.sum():", hist.sum(), "img.shape:", img.shape)

# 히스토그램 그래프를 화면에 표시
plt.show()

# OpenCV 창을 유지하기 위해 키 입력 대기
cv2.waitKey()
cv2.destroyAllWindows()  # 모든 창을 닫음
```
