```python
import cv2
import numpy as np
import matplotlib.pylab as plt

#--① 그레이 스케일로 영상 읽기
img = cv2.imread('../img/abnormal.jpg', cv2.IMREAD_GRAYSCALE)  # 이미지를 그레이 스케일로 읽어옵니다.

#--② 직접 연산한 정규화
img_f = img.astype(np.float32)  # 이미지를 float32 형식으로 변환합니다.
# (최소값 - 이미지) / (최대값 - 최소값) * 255: 이미지의 값을 0-255 범위로 정규화합니다.
img_norm = ((img_f - img_f.min()) * (255) / (img_f.max() - img_f.min()))
img_norm = img_norm.astype(np.uint8)  # 정규화된 이미지를 uint8 형식으로 변환합니다.

#--③ OpenCV API를 이용한 정규화
img_norm2 = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)  # OpenCV의 normalize 함수를 이용해 정규화합니다.

#--④ 히스토그램 계산
hist = cv2.calcHist([img], [0], None, [256], [0, 255])  # 원본 이미지의 히스토그램을 계산합니다.
hist_norm = cv2.calcHist([img_norm], [0], None, [256], [0, 255])  # 직접 정규화한 이미지의 히스토그램을 계산합니다.
hist_norm2 = cv2.calcHist([img_norm2], [0], None, [256], [0, 255])  # OpenCV로 정규화한 이미지의 히스토그램을 계산합니다.

#--⑤ 이미지와 히스토그램 출력
cv2.imshow('Before', img)  # 원본 이미지를 화면에 표시합니다.
cv2.imshow('Manual', img_norm)  # 직접 정규화한 이미지를 화면에 표시합니다.
cv2.imshow('cv2.normalize()', img_norm2)  # OpenCV로 정규화한 이미지를 화면에 표시합니다.

#--⑥ 히스토그램을 플로팅하여 출력
hists = {'Before': hist, 'Manual': hist_norm, 'cv2.normalize()': hist_norm2}
for i, (k, v) in enumerate(hists.items()):  # 각 히스토그램을 반복하여 플로팅합니다.
    plt.subplot(1, 3, i + 1)  # 1행 3열의 서브플롯을 설정합니다.
    plt.title(k)  # 서브플롯의 제목을 설정합니다.
    plt.plot(v)  # 히스토그램을 플로팅합니다.
plt.show()  # 플로팅된 히스토그램을 화면에 표시합니다.

# OpenCV 창을 유지하기 위해 키 입력 대기
cv2.waitKey()
cv2.destroyAllWindows()  # 모든 창을 닫음
```
