```python
import cv2
import numpy as np
import matplotlib.pylab as plt

#--① 이미지 읽기 및 출력
img = cv2.imread('../img/mountain.jpg')  # 이미지를 BGR 형식으로 읽어옵니다.
cv2.imshow('img', img)  # 이미지를 화면에 출력합니다.

#--② 히스토그램 계산 및 그리기
channels = cv2.split(img)  # 이미지의 채널(B, G, R)을 분리합니다.
colors = ('b', 'g', 'r')  # 각 채널에 해당하는 색상(B, G, R)을 지정합니다.
for (ch, color) in zip(channels, colors):  # 각 채널과 색상을 반복문을 통해 처리합니다.
    hist = cv2.calcHist([ch], [0], None, [256], [0, 255])  # 각 채널의 히스토그램을 계산합니다.
    plt.plot(hist, color=color)  # 해당 채널의 히스토그램을 지정된 색상으로 그립니다.
plt.show()  # 히스토그램을 화면에 출력합니다.

# OpenCV 창을 유지하기 위해 키 입력 대기
cv2.waitKey()
cv2.destroyAllWindows()  # 모든 창을 닫음
```
