```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

# 침식과 팽창 연산을 수행하는 함수
def erode_dilate():
    # 이미지를 그레이스케일로 불러오기
    src = cv2.imread('milkdrop.bmp', cv2.IMREAD_GRAYSCALE)

    # 이미지 로드 실패 시 오류 메시지 출력 후 함수 종료
    if src is None:
        print('Image load failed!')
        return

    # 이진화: Otsu의 방법을 사용하여 이진 이미지 생성
    _, src_bin = cv2.threshold(src, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    # 침식 연산 수행
    dst1 = cv2.erode(src_bin, None)
    
    # 팽창 연산 수행
    dst2 = cv2.dilate(src_bin, None)

    # 결과 이미지 표시
    plt.subplot(221), plt.axis('off'), plt.imshow(src, 'gray'), plt.title('src')  # 원본 이미지
    plt.subplot(222), plt.axis('off'), plt.imshow(src_bin, 'gray'), plt.title('src_bin')  # 이진화된 이미지
    plt.subplot(223), plt.axis('off'), plt.imshow(dst1, 'gray'), plt.title('erode')  # 침식 연산 결과
    plt.subplot(224), plt.axis('off'), plt.imshow(dst2, 'gray'), plt.title('dilate')  # 팽창 연산 결과
    # plt 이미지 파일로 저장
    plt.savefig('C:/Users/winrr/Pictures/Screenshots/pfile/morpgology1.jpg', bbox_inches='tight')  
    plt.show()
 

# 열림과 닫힘 연산을 수행하는 함수
def open_close():
    # 이미지를 그레이스케일로 불러오기
    src = cv2.imread('milkdrop.bmp', cv2.IMREAD_GRAYSCALE)

    # 이미지 로드 실패 시 오류 메시지 출력 후 함수 종료
    if src is None:
        print('Image load failed!')
        return

    # 이진화: Otsu의 방법을 사용하여 이진 이미지 생성
    _, src_bin = cv2.threshold(src, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    # 열림 연산 수행 (침식 후 팽창)
    dst1 = cv2.morphologyEx(src_bin, cv2.MORPH_OPEN, None)
    
    # 닫힘 연산 수행 (팽창 후 침식)
    dst2 = cv2.morphologyEx(src_bin, cv2.MORPH_CLOSE, None)

    # 결과 이미지 표시
    plt.subplot(221), plt.axis('off'), plt.imshow(src, 'gray'), plt.title('src')  # 원본 이미지
    plt.subplot(222), plt.axis('off'), plt.imshow(src_bin, 'gray'), plt.title('src_bin')  # 이진화된 이미지
    plt.subplot(223), plt.axis('off'), plt.imshow(dst1, 'gray'), plt.title('open')  # 열림 연산 결과
    plt.subplot(224), plt.axis('off'), plt.imshow(dst2, 'gray'), plt.title('close')  # 닫힘 연산 결과
    # plt 이미지 파일로 저장
    plt.savefig('C:/Users/winrr/Pictures/Screenshots/pfile/morpgology2.jpg', bbox_inches='tight') 
    plt.show()
    
# 스크립트 실행 부분
if __name__ == '__main__':
    erode_dilate()  # 침식 및 팽창 연산 함수 호출
    open_close()  # 열림 및 닫힘 연산 함수 호출
```
