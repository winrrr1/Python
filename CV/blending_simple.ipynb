```python
import cv2
import numpy as np
import matplotlib.pylab as plt

# ---① 연산에 사용할 이미지 읽기
# 두 개의 이미지를 읽어옴
img1 = cv2.imread('../img/wing_wall.jpg')
img2 = cv2.imread('../img/yate.jpg')

# ---② 이미지 덧셈
# 두 이미지를 더하는 연산
img3 = img1 + img2  # NumPy를 이용한 이미지 덧셈
img4 = cv2.add(img1, img2) # OpenCV의 add 함수를 이용한 이미지 덧셈

# 이미지들을 딕셔너리에 저장
imgs = {'img1':img1, 'img2':img2, 'img1+img2': img3, 'cv.add(img1, img2)': img4}

# ---③ 이미지 출력
# 각 이미지를 2x2 형태로 출력
for i, (k, v) in enumerate(imgs.items()):
    plt.subplot(2, 2, i + 1)  # 2x2 격자에서 이미지의 위치를 설정
    plt.imshow(v[:,:,::-1])  # 이미지를 BGR에서 RGB로 변환하여 표시
    plt.title(k)  # 이미지 제목 설정
    plt.xticks([]); plt.yticks([])  # x축과 y축의 눈금 표시 제거

plt.show()  # 모든 이미지를 화면에 표시
```
