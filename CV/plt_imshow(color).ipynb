```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#    
img = cv2.imread('../img/girl.jpg')

# OpenCV  BGR    RGB 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#  
plt.figure(figsize=(12, 8))

plt.subplot(3, 3, 1)
plt.title('Original RGB')
plt.imshow(img_rgb)
plt.axis('off')

# BGR      
channels = [
    (0, 1, 2),  # BGR ( ,    )
    (1, 0, 2),  # G, B, R
    (1, 2, 0),  # G, R, B
    (2, 0, 1),  # B, R, G
    (0, 2, 1),  # B, R, G
    (2, 1, 0)   # R, G, B (  )
]

titles = [
    'BGR (Incorrect)',
    'G, B, R',
    'G, R, B',
    'B, R, G',
    'B, R, G',
    'R, G, B (Correct)'
]

for i, ch in enumerate(channels):
    plt.subplot(3, 3, i + 2)
    plt.title(titles[i])
    plt.imshow(img[:, :, ch])
    plt.axis('off')

plt.show()
```
