```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 이미지 로드 및 변환
img = cv2.imread('../img/girl.jpg')

# OpenCV는 이미지를 BGR 포맷으로 읽기 때문에 RGB로 변환
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 원본 컬러
plt.figure(figsize=(12, 8))

plt.subplot(3, 3, 1)
plt.title('Original RGB')
plt.imshow(img_rgb)
plt.axis('off')

# BGR 채널을 다른 순서로 표시하여 색상 변형
channels = [
    (0, 1, 2),  # BGR (잘못된 순서, 원본이 아닌 색상으로 표시됨)
    (1, 0, 2),  # G, B, R
    (1, 2, 0),  # G, R, B
    (2, 0, 1),  # B, R, G
    (0, 2, 1),  # B, R, G
    (2, 1, 0)   # R, G, B (원본과 같은 색상)
]

titles = [
    'BGR (Incorrect)',
    'G, B, R',
    'G, R, B',
    'B, R, G',
    'B, R, G',
    'R, G, B (Correct)'
]

for i, ch in enumerate(channels):
    plt.subplot(3, 3, i + 2)
    plt.title(titles[i])
    plt.imshow(img[:, :, ch])
    plt.axis('off')

plt.show()
```
