```python
import cv2
import numpy as np
import matplotlib.pylab as plt

#--① 이미지 읽기 및 HSV 색상 공간으로 변환
img = cv2.imread("../img/cube.jpg")  # 이미지를 읽어옴
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  # BGR 색상 공간에서 HSV 색상 공간으로 변환

#--② 색상 범위 정의
# HSV 색상 공간에서 색상 범위를 정의
blue1 = np.array([90, 50, 50])       # 파란색의 하한
blue2 = np.array([120, 255, 255])    # 파란색의 상한
green1 = np.array([45, 50, 50])      # 초록색의 하한
green2 = np.array([75, 255, 255])    # 초록색의 상한
red1 = np.array([0, 50, 50])         # 빨간색의 하한 (첫 번째 범위)
red2 = np.array([15, 255, 255])      # 빨간색의 상한 (첫 번째 범위)
red3 = np.array([165, 50, 50])       # 빨간색의 하한 (두 번째 범위)
red4 = np.array([180, 255, 255])     # 빨간색의 상한 (두 번째 범위)
yellow1 = np.array([20, 50, 50])     # 노란색의 하한
yellow2 = np.array([35, 255, 255])   # 노란색의 상한

# --③ 색상 범위에 따라 마스크 생성
# 특정 색상의 범위에 해당하는 부분을 흰색으로, 나머지는 검정색으로 마스크를 만듦
mask_blue = cv2.inRange(hsv, blue1, blue2)    # 파란색 마스크
mask_green = cv2.inRange(hsv, green1, green2) # 초록색 마스크
mask_red = cv2.inRange(hsv, red1, red2)       # 빨간색 첫 번째 범위 마스크
mask_red2 = cv2.inRange(hsv, red3, red4)      # 빨간색 두 번째 범위 마스크
mask_yellow = cv2.inRange(hsv, yellow1, yellow2) # 노란색 마스크

#--④ 색상별로 원본 이미지에서 색상만 추출
# 마스크를 이용해 해당 색상만 추출
res_blue = cv2.bitwise_and(img, img, mask=mask_blue)    # 파란색 추출
res_green = cv2.bitwise_and(img, img, mask=mask_green)  # 초록색 추출
res_red1 = cv2.bitwise_and(img, img, mask=mask_red)     # 빨간색 첫 번째 범위 추출
res_red2 = cv2.bitwise_and(img, img, mask=mask_red2)    # 빨간색 두 번째 범위 추출
res_red = cv2.bitwise_or(res_red1, res_red2)           # 두 범위의 빨간색을 합침
res_yellow = cv2.bitwise_and(img, img, mask=mask_yellow) # 노란색 추출

#--⑤ 결과 출력
# 원본 이미지와 색상별 추출된 이미지를 출력
imgs = {'original': img, 'blue': res_blue, 'green': res_green, 
        'red': res_red, 'yellow': res_yellow}
for i, (k, v) in enumerate(imgs.items()):
    plt.subplot(2, 3, i + 1)         # 2행 3열의 서브플롯을 설정
    plt.title(k)                    # 각 서브플롯의 제목 설정
    plt.imshow(v[:, :, ::-1])       # OpenCV는 BGR 순서로 읽으므로 RGB로 변환하여 출력
    plt.xticks([]); plt.yticks([])  # x, y축의 눈금을 숨김
plt.show()                          # 결과를 화면에 표시
```
