```python
import numpy as np
import cv2

# 직선을 그리는 함수 정의
def drawLines():
    img = np.full((400, 400, 3), 255, np.uint8)  # 400x400 크기의 흰색 배경 이미지 생성

    # 다양한 두께와 색상의 직선 그리기
    cv2.line(img, (50, 50), (200, 50), (0, 0, 255))  # 빨간색 직선
    cv2.line(img, (50, 100), (200, 100), (255, 0, 255), 3)  # 두께 3의 보라색 직선
    cv2.line(img, (50, 150), (200, 150), (255, 0, 0), 10)  # 두께 10의 파란색 직선

    # 다양한 선 유형으로 직선 그리기
    cv2.line(img, (250, 50), (350, 100), (0, 0, 255), 1, cv2.LINE_4)  # LINE_4 유형의 빨간색 직선
    cv2.line(img, (250, 70), (350, 120), (255, 0, 255), 1, cv2.LINE_8)  # LINE_8 유형의 보라색 직선
    cv2.line(img, (250, 90), (350, 140), (255, 0, 0), 1, cv2.LINE_AA)  # LINE_AA(안티앨리어스) 유형의 파란색 직선

    # 화살표가 포함된 직선 그리기
    cv2.arrowedLine(img, (50, 200), (150, 200), (0, 0, 255), 1)  # 빨간색 화살표 직선
    cv2.arrowedLine(img, (50, 250), (350, 250), (255, 0, 255), 1)  # 보라색 화살표 직선
    cv2.arrowedLine(img, (50, 300), (350, 300), (255, 0, 0), 1, cv2.LINE_8, 0, 0.05)  # 파란색 화살표 직선, 머리 크기 조정

    # 다양한 마커 그리기
    cv2.drawMarker(img, (50, 350), (0, 0, 255), cv2.MARKER_CROSS)  # 십자가 마커
    cv2.drawMarker(img, (100, 350), (0, 0, 255), cv2.MARKER_TILTED_CROSS)  # 기울어진 십자가 마커
    cv2.drawMarker(img, (150, 350), (0, 0, 255), cv2.MARKER_STAR)  # 별 모양 마커
    cv2.drawMarker(img, (200, 350), (0, 0, 255), cv2.MARKER_DIAMOND)  # 다이아몬드 모양 마커
    cv2.drawMarker(img, (250, 350), (0, 0, 255), cv2.MARKER_SQUARE)  # 사각형 마커
    cv2.drawMarker(img, (300, 350), (0, 0, 255), cv2.MARKER_TRIANGLE_UP)  # 위쪽 삼각형 마커
    cv2.drawMarker(img, (350, 350), (0, 0, 255), cv2.MARKER_TRIANGLE_DOWN)  # 아래쪽 삼각형 마커

    # 이미지를 화면에 표시하고 창 닫기
    cv2.imshow("img", img)
    cv2.waitKey()
    cv2.destroyAllWindows()

# 다각형을 그리는 함수 정의
def drawPolys():
    img = np.full((400, 400, 3), 255, np.uint8)  # 400x400 크기의 흰색 배경 이미지 생성

    # 사각형 그리기
    cv2.rectangle(img, (50, 50), (150, 100), (0, 0, 255), 2)  # 테두리만 있는 빨간색 사각형
    cv2.rectangle(img, (50, 150), (150, 200), (0, 0, 128), -1)  # 채워진 진한 빨간색 사각형

    # 원 그리기
    cv2.circle(img, (300, 120), 30, (255, 255, 0), -1, cv2.LINE_AA)  # 채워진 노란색 원
    cv2.circle(img, (300, 120), 60, (255, 0, 0), 3, cv2.LINE_AA)  # 테두리만 있는 파란색 원

    # 타원 그리기
    cv2.ellipse(img, (120, 300), (60, 30), 20, 0, 270, (255, 255, 0), cv2.FILLED, cv2.LINE_AA)  # 부분적으로 채워진 노란색 타원
    cv2.ellipse(img, (120, 300), (100, 50), 20, 0, 360, (0, 255, 0), 2, cv2.LINE_AA)  # 테두리만 있는 녹색 타원

    # 다각형 그리기
    pts = np.array([[250, 250], [300, 250], [300, 300], [350, 300], [350, 350], [250, 350]])  # 다각형의 꼭지점 좌표
    cv2.polylines(img, [pts], True, (255, 0, 255), 2)  # 보라색 테두리의 다각형 그리기

    # 이미지를 화면에 표시하고 창 닫기
    cv2.imshow("img", img)
    cv2.waitKey()
    cv2.destroyAllWindows()

# 텍스트를 그리는 함수 정의 (여러 폰트 스타일 사용)
def drawText1():
    img = np.full((500, 800, 3), 255, np.uint8)  # 500x800 크기의 흰색 배경 이미지 생성

    # 다양한 폰트 스타일의 텍스트 그리기
    cv2.putText(img, "FONT_HERSHEY_SIMPLEX", (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255))  # 간단한 폰트
    cv2.putText(img, "FONT_HERSHEY_PLAIN", (20, 100), cv2.FONT_HERSHEY_PLAIN, 1, (0, 0, 255))  # 단순한 폰트
    cv2.putText(img, "FONT_HERSHEY_DUPLEX", (20, 150), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 255))  # 이중 폰트
    cv2.putText(img, "FONT_HERSHEY_COMPLEX", (20, 200), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0))  # 복잡한 폰트
    cv2.putText(img, "FONT_HERSHEY_TRIPLEX", (20, 250), cv2.FONT_HERSHEY_TRIPLEX, 1, (255, 0, 0))  # 삼중 폰트
    cv2.putText(img, "FONT_HERSHEY_COMPLEX_SMALL", (20, 300), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255, 0, 0))  # 작은 복잡한 폰트
    cv2.putText(img, "FONT_HERSHEY_SCRIPT_SIMPLEX", (20, 350), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 1, (255, 0, 255))  # 간단한 스크립트 폰트
    cv2.putText(img, "FONT_HERSHEY_SCRIPT_COMPLEX", (20, 400), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 1, (255, 0, 255))  # 복잡한 스크립트 폰트
    cv2.putText(img, "FONT_HERSHEY_COMPLEX | FONT_ITALIC", (20, 450), cv2.FONT_HERSHEY_COMPLEX | cv2.FONT_ITALIC, 1, (255, 0, 0))  # 이탤릭 복잡한 폰트

    # 이미지를 화면에 표시하고 창 닫기
    cv2.imshow("img", img)
    cv2.waitKey()
    cv2.destroyAllWindows()

# 텍스트를 가운데 정렬하여 그리는 함수 정의
def drawText2():
    img = np.full((200, 640, 3), 255, np.uint8)  # 200x640 크기의 흰색 배경 이미지 생성

    text = "Hello, OpenCV"  # 표시할 텍스트
    fontFace = cv2.FONT_HERSHEY_TRIPLEX  # 폰트 스타일 설정
    fontScale = 2.0  # 폰트 크기 설정
    thickness = 1  # 폰트 두께 설정

    # 텍스트 크기 계산
    sizeText, _ = cv2.getTextSize(text, fontFace, fontScale, thickness)

    # 텍스트를 이미지 중앙에 위치시키기 위한 좌표 계산
    org = ((img.shape[1] - sizeText[0]) // 2, (img.shape[0] + sizeText[1]) // 2)
    cv2.putText(img, text, org, fontFace, fontScale, (255, 0, 0), thickness)  # 텍스트 그리기
    cv2.rectangle(img, org, (org[0] + sizeText[0], org[1] - sizeText[1]), (0, 255, 0), 1)  # 텍스트 주위에 사각형 그리기

    # 이미지를 화면에 표시하고 창 닫기
    cv2.imshow("img", img)
    cv2.waitKey()
    cv2.destroyAllWindows()

# 메인 함수에서 정의된 함수들을 호출하여 실행
if __name__ == '__main__':
    drawLines()  # 직선 그리기 함수 실행
    drawPolys()  # 다각형 그리기 함수 실행
    drawText1()  # 텍스트 그리기 함수 1 실행
    drawText2()  # 텍스트 그리기 함수 2 실행
```
