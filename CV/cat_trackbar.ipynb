```python
import cv2
import numpy as np

win_name = 'cat_Trackbar'  # 창 제목 설정

# 이미지 파일 읽기
img = cv2.imread('../img/cat1.jpg')

# # 이미지가 제대로 로드되었는지 확인
# if img is None:
#     print('이미지를 불러올 수 없습니다. 파일 경로를 확인하십시오.')
#     exit()

# 이미지 복사본 생성
img_copy = img.copy()

# 초기 이미지를 창에 표시
cv2.imshow(win_name, img)

# 트랙바 이벤트 처리 함수 정의 ---①
def onChange(x):
    # 'R', 'G', 'B' 각 트랙바의 현재 위치 값 읽기 ---③
    r = cv2.getTrackbarPos('R', win_name)
    g = cv2.getTrackbarPos('G', win_name)
    b = cv2.getTrackbarPos('B', win_name)
    
    # 현재 RGB 값 출력
    print(r, g, b)
    
    # 색상을 적용한 이미지 복사본을 업데이트
    img_copy[:] = [b, g, r]  # OpenCV는 BGR 포맷을 사용하므로 R, G, B 순서로 적용
    
    # 원본 이미지와 색상 조정 이미지를 결합
    blended_img = cv2.addWeighted(img, 0.5, img_copy, 0.5, 0)
    
    # 업데이트된 이미지를 창에 다시 표시
    cv2.imshow(win_name, blended_img)

# 트랙바 생성 ---⑤
# 'R', 'G', 'B' 각각의 트랙바를 생성하고, 최대 값 255, 초기 값 0, onChange 함수 연결
cv2.createTrackbar('R', win_name, 0, 255, onChange)
cv2.createTrackbar('G', win_name, 0, 255, onChange)
cv2.createTrackbar('B', win_name, 0, 255, onChange)

# 초기 트랙바 값 설정 ---⑥
cv2.setTrackbarPos('R', win_name, 255)
cv2.setTrackbarPos('G', win_name, 255)
cv2.setTrackbarPos('B', win_name, 255)

# 무한 루프를 돌며 키 입력 대기
while True:
    if cv2.waitKey(1) & 0xFF == 27:  # ESC 키를 눌러 종료
        break

# 모든 윈도우 닫기
cv2.destroyAllWindows()
```
